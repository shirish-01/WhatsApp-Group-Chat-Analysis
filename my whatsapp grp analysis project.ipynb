```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("meme war.csv",header=None,names=["chats"])
```

```python
df
```

```python
df.info()
df.describe()
```

```python
new= df["chats"].str.split("-", expand = True)
df["time"] = new[0]
df["chats"] = new[1]
```

```python
df.drop(index=0)
```

```python
df =df.drop(index=0)
df
```

```python
namesplit = df["chats"].str.split(":",expand=True)
df["names"] = namesplit[0]
df["messages"] = namesplit[1]
```

```python
df
```

```python
#not for time pass
participants = 11
dt = df["names"].value_counts().head(participants)
```

```python
color = ["#D3D3D3","#228B22", "#FF0000", "#00FF00", "#FFFF00", "#0000FF", "#FFC0CB", "#696969", "#A52A2A", "#FFA500" ,"#800080"]
dt.plot(x=df["names"].value_counts().head(participants),kind = "pie",radius=3,colors= color,)

plt.grid()
plt.legend()
```

```python
dt.plot(x=df["names"].value_counts().head(participants),kind = "bar",width = 0.5,)
plt.xlabel("participants name",color="green",size =15)
plt.ylabel("no of messages given by each one",color ="green",size=15)
plt.grid(axis='y')
plt.title("our grp analysis",size="20")
```
